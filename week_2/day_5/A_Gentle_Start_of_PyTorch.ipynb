##  Section 1: Tensor

In this section, we introduce Tensor and its basic operations.  

This section aims to help you get familiar with the following key questions about tensor:
* What is a tensor?
* How to generate a tensor?
* Attributes of a tensor?
* Indexing, slicing, and joining of tensor?

```python
# import pytorch library and other necessary library(ies)

import torch
import numpy as np
```

### Q1 - What is a tensor?

In PyTorch, we use tensors to encode the inputs and outputs of a model, as well as the modelâ€™s parameters.

Tensors are a specialized data structure that are very similar to arrays and matrices. We know that, a matrix is a 2-D grid of numbers, then, we can view tensor as a generalized matrix. It can be a 1-D matrix (i.e. vector, which is a one-row grid of numbers), 2-D matrix, or 3-D matrix (a cube of numbers), or a higher dimensional matrix.

Compared to ndarrays in NumPy, tensors can run on GPUs or other hardware accelerators, and can be optimized during the training.

### Q2 - How to generate a tensor?

```python
'''
A tensor can be created:
  - Directly from data
  - From a NumPy array
  - With constant values
  - With random values
'''

# create from data

data = [[1, 2],[3, 4]]
tensor_data = torch.tensor(data)
print("tensor created directly from data: \n", tensor_data)
```

```python
# create from Numpy array

np_array = np.array(data)
tensor_np = torch.from_numpy(np_array)
print("tensor created from Numpy array: \n", tensor_np)
```

```python
# create with constant values

shape_const = (2,3) # specify the shape of tensor

ones_tensor = torch.ones(shape_const) # all-one tensor
zeros_tensor = torch.zeros(shape_const) # all-zero tensor

print("All-one tensor: \n", ones_tensor)
print("\n All-zero Tensor: \n", zeros_tensor)
```

```python
# create with random values

shape_rand = (2,3) #specify the shape of tensor
rand_tensor = torch.rand(shape_rand)
print("Random tensor: \n", rand_tensor)
```

### Q3 - Attributes of a tensor?

Tensor attributes describe their shape, datatype, and the device on which they are stored.

```python
# take rand_tensor created before as an example, print its attributes

print("Shape of rand_tensor: ", rand_tensor.shape)
print("Datatype of rand_tensor: ", rand_tensor.dtype)
print("Device rand_tensor is stored on: ", rand_tensor.device)
```

```python
# You can create a certain type of a tensor

data = [[1, 2],[3, 4]]
tensor_data = torch.tensor(data) # default tensor type for this case would be LongTensor
float_tensor_data = torch.FloatTensor(data) # but can create a float tensor for this case

print("Datatype of tensor_data: ", tensor_data.dtype)
print("Datatype of float_tensor_data: ", float_tensor_data.dtype)
```

```python
# You can change the datatype of a tensor

print("Datatype of ones_tensor before change: ", ones_tensor.dtype)
ones_tensor = ones_tensor.type(torch.LongTensor) # change the type of ones_tensor from float32 to long integer (i.e. int64)
print("Datatype of ones_tensor after change: ", ones_tensor.dtype)
```

### Q4 - Indexing, slicing, and joining of tensor?

```python
# create example tensors, will do indexing, slicing, joining with the example tensors

data_1 = [[1, 2],[3, 4],[5, 6]]
tensor_1 = torch.tensor(data_1)
print("tensor 1: \n", tensor_1)

data_2 = [[7, 8],[9, 10]]
tensor_2 = torch.tensor(data_2)
print("tensor 2: \n", tensor_2)
```

```python
# indexing: get an element/ a row/ a colum/ some rows of tensor1

print("First element: ", tensor_1[0,0])
print('First row: ', tensor_1[0, :])
print('First column: ', tensor_1[:, 0])
print('Last column:', tensor_1[..., -1])
print('First 2 rows: \n', tensor_1[:2,])
```

```python
# slicing: change a column of tensor_1

tensor_1[:,1] = 0
print(tensor_1)
```

```python
# joining: concatenate

concat_tensor = torch.cat([tensor_1, tensor_2])
print("concatenation of tensor_1 and tensor_2: \n", concat_tensor)
```

### Q5 - Arithmetic operations of tensor?

```python
'''
Some frequently used arithmetic operations of tensor:
  - add 2 tensors
  - multiplication
  - element-wise product
'''

# create example tensors, will explain operations with the example tensors

data_3 = [[1, 2],[3, 4]]
tensor_3 = torch.tensor(data_3)
print("tensor 3: \n", data_3)

data_4 = [[5, 6],[7, 8]]
tensor_4 = torch.tensor(data_4)
print("tensor 4: \n", tensor_4)
```

```python
# This computes add operation between two tensores

add = tensor_3 + tensor_4
print("Adding tensor_3 and tensor_4: \n", add)
```

```python

# This computes the matrix multiplication between two tensors.
#  mul_1 and mul_2 are just two different ways to do it and will have the same value
mul_1 = tensor_3 @ tensor_4
mul_2 = tensor_3.matmul(tensor_4)
print("Multiplying tensor_3 and tensor_4 with @: \n", mul_1)
print("\n Multiplying tensor_3 and tensor_4 with .matmul: \n", mul_2)

```

```python
# This computes the element-wise product.
# ele_mul_1 and ele_mul_2 are just two different ways to do it and will have the same value

ele_mul_1 = tensor_3 * tensor_4
ele_mul_2 = tensor_3.mul(tensor_4)

print("Element-wise multiplication of tensor_3 and tensor_4 with *: \n", ele_mul_1)
print("\n Element-wise multiplication of  tensor_3 and tensor_4 with .mul: \n", ele_mul_2)


```

## Section 2: A Toy Neural Network Example

In this section, we implement a simple 2-layer NN classifier (with linear layers only), and appy it to a synthetic dataset.

This section aims to provide a simple example and help you get familiar with the following key steps in Neural Network training with PyTorch:
* How to build NN model with PyTorch pre-defined layers?
* How to instantiate the model, and define the optimizer and loss function?
* How to train the model?

```python
# import neccessary libraries

import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# define hyper-parameters

toy_learning_rate = 0.02
toy_epoch_num = 29
```

```python
# generate synthetic dataset and visualize it

torch.manual_seed(10) # fix random seed for reproducibility

n_data = torch.ones(200, 2)
x0 = torch.normal(2*n_data, 1)      # class0 x data (tensor), shape=(200, 2)
y0 = torch.zeros(200)               # class0 y data (tensor), shape=(200, 1)
x1 = torch.normal(-2*n_data, 1)     # class1 x data (tensor), shape=(200, 2)
y1 = torch.ones(200)                # class1 y data (tensor), shape=(200, 1)
x = torch.cat((x0, x1), 0).type(torch.FloatTensor)  # shape (400, 2) FloatTensor = 32-bit floating
y = torch.cat((y0, y1), ).type(torch.LongTensor)    # shape (400,) LongTensor = 64-bit integer

# visualize the generated data
plt.scatter(x.data.numpy()[:, 0], x.data.numpy()[:, 1], c=y.data.numpy(), s=100, lw=0, cmap='RdYlGn')
plt.show()
```

### Q1 - How to build NN model with PyTorch pre-defined layers?

```python
# A two layer neural network model

class toy_2_layer_NN(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(toy_2_layer_NN, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)   # a linear layer with weight matrix of dimension n_feature * n_hidden
        self.out = torch.nn.Linear(n_hidden, n_output)   # a linear layer with weight matrix of dimension n_hidden * n_output

    def forward(self, x):
        # Forward Pass of this model
        # The model architecture is: linear layer -> relu activation -> linear layer
        x = F.relu(self.hidden(x))      # non-linear activation function for hidden layer
        x = self.out(x)
        return x
```

### Q2 - How to instantiate the model, and define the optimizer and loss function?

```python
# instantiate the model, define the optimizer and loss function

toy_net = toy_2_layer_NN(n_feature=2, n_hidden=10, n_output=2)     # define the network
toy_optimizer = torch.optim.SGD(toy_net.parameters(), lr=toy_learning_rate) # define the optimizer
toy_loss_func = torch.nn.CrossEntropyLoss()  # the target label is NOT an one-hotted

print(toy_net) # print model architecture
```

### Q3 - How to train the model?

```python
'''
Training:
  you would be able to see how the prediction results and the training accuracy get updated in this
  toy example, we aim to help you get familiar with how a neural network is implemented and trained,
  therefore, we do not include a test stage in this example
'''

plt.ion()   # a function for plotting

for epoch in range(toy_epoch_num):

    toy_optimizer.zero_grad()   # clear gradients for next train
    out = toy_net(x)                 # input x and predict based on x
    toy_loss = toy_loss_func(out, y)     # must be (1. nn output, 2. target), the target label is NOT one-hotted
    toy_loss.backward()         # backpropagation, compute gradients
    toy_optimizer.step()        # apply gradients to optimize the parameters

    if epoch % 4 == 1:
        # plot and show learning process for every 4 epochs

        # make predictions
        prediction = torch.max(out, 1)[1]
        pred_y = prediction.data.numpy()
        target_y = y.data.numpy()

        # plot
        plt.cla()
        plt.scatter(x.data.numpy()[:, 0], x.data.numpy()[:, 1], c=pred_y, s=100, lw=0, cmap='RdYlGn') # plot
        accuracy = float((pred_y == target_y).astype(int).sum()) / float(target_y.size) # compute accuracy
        plt.text(1.5, -4, 'Accuracy=%.2f' % accuracy, fontdict={'size': 20, 'color':  'red'})
        plt.pause(0.1)

plt.ioff()
plt.show()
```
